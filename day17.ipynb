```python
x_min, x_max = 235, 259
y_min, y_max = -118, -62
```

```python
itr = 0
maxmaxy = 0
num_success_inits = 0
for u0 in tqdm(range(0,x_max+1)):
    for v0 in range(y_min,abs(y_min)):
        u,v = u0,v0
        x, y = 0, 0
        maxy = 0
        success = False
        for itr in range(300):
            x += u
            y += v
            maxy = max(maxy,y)
            u -= np.sign(u)
            v -= 1
            if x_min <= x <= x_max and y_min <= y <= y_max:
                success = True
                num_success_inits += 1
                break
            if x > x_max:
                break
        if success:
            maxmaxy = max(maxmaxy,maxy)
            print(f"success with {u0 = }, {u = }, {v0 = }, {v = }, {itr = }, {x = }, {y = }")
print(f"{maxmaxy = }, {num_success_inits = }")
```

```python
puzzle.answer_a = maxmaxy
```
