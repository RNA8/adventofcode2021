```python
import os
import numpy as np
import aocd
from aocd.models import Puzzle
from aocd import submit
```

```python
current_day = 10
current_year = 2021
puzzle = Puzzle(year=current_year, day=current_day)
puzzle
```

```python
data = puzzle.input_data.splitlines()
data
```

# Part 1

```python
matches = {"]":"[",
           ">":"<",
           "}":"{",
           ")":"("}
scores = {')': 3,
          ']': 57,
          '}': 1197,
          '>': 25137}
```

```python
from collections import deque
score = 0
for line in data:
    stack = deque()
    for char in line:
        if char in '({[<':
            stack.append(char)
        else:
            if matches[char] == stack[-1]:
                stack.pop()
            else: # mismatch
                score += scores[char]
                break
```

```python
score
```

```python
puzzle.answer_a = score
```

# Part 2

```python
closing_scores = {"(": 1,
"[": 2,
"{": 3,
"<": 4}

from collections import deque
scores = []
for line in data:
    stack = deque()
    score = 0
    mismatch = False
    for char in line:
        if char in '({[<':
            stack.append(char)
        else:
            if matches[char] == stack[-1]:
                stack.pop()
            else: # mismatch - continue
                mismatch = True
                break
    if mismatch:
        continue
    # pop the remaining things in the stack
    while stack:
        score = score * 5 + closing_scores[stack[-1]]
        stack.pop()
    scores.append(score)
int(np.median(scores))
```
